```python
import math

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    return a / b

def power(a, b):
    return a ** b

def square_root(x):
    return math.sqrt(x)

def factorial(n):
    if int(n) != n or n < 0:
        raise ValueError("Factorial only for non-negative integers")
    n = int(n)
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact

def quick_report(a, b):
    return {
        "add": a + b,
        "subtract": a - b,
        "multiply": a * b,
        "divide": a / b if b != 0 else "undefined"
    }
```

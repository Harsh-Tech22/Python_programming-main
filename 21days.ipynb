***What is Encapsullation in Python***

```python
class student:
    def __init__(self):
        self .__marks = 90

    def show_marks(self):
        return self .__marks
s = student()
print(s.show_marks())
```

***What is Abstraction in Python***

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        return "car starts with key"

c = Car()
print(c.start())
```

***What is Inheritance in Python***

```python
class vehicle:
    def start(self):
        return "vehicle starts"
class car(vehicle):
    pass
c = car()
print(c.start())
```

***What is Polymorphism in Python***

```python
class dog:
    def sound(self):
        return "bark"
class cat:
    def sound(self):
        return "meow"
d = dog()
c = cat()
print(d.sound())
print(c.sound())
```

***Q3. Static Method***

```python
class MathHelper:
    @staticmethod
    def is_even(num):
        return num % 2 == 0
print(MathHelper.is_even(8))
print(MathHelper.is_even(11))
    
```

***Public and Private Data Members***

```python
class BankAccount:
    def __init__(self,name,balance):
        self.name = name
        self.__balance = balance

    def deposite(self,amount):

        if amount > 0:
            self.__balance += amount
            print(f"Deposited:{amount}")
        else:
            print("No Blanace")
        
    def withdraw(self,amount):
        if amount > self.__balance:
            print("INSUFFICIENT BALANCE")
        elif amount <= 0:
            print("Withdrawl amount should be in positive")
        else:
            self.__balance -= amount
            print(f"Withdrawn:{amount}")
    def show_balance(self):
        print(f"Account Holder:{self.name}")
        print(f"Current Balance:{self.__balance}")

my_account = BankAccount("Harsh Kumar", 5000)
my_account.show_balance()
my_account.deposite(500)
my_account.withdraw(200)
my_account.withdraw(2000)
my_account.show_balance()
```

***Magic Methods / Operator Overloading***

```python
class product:
    def __init__(self,name,price):
        self.name = name
        self.price = price
    def __str__(self):
        return f"product: {self.name}, Price:{self.price}"
    def __add__(self,other):
        return self.price + other.price
    def __eq__(self,other):
        return self.price == other.price
```

```python
p1 = product("laptop", 10000)
p2 = product("phone", 50000)
p3 = product("Tablet", 70000)

print(p1)
print(p1 + p2)
print(p1 == p3)
print(p2 == p1)
```

***Instance method vs Class method***

```python
class student:
    def __init__(self,name):
        self.name = name

        def show(self):
            print(self.name)

```

```python
class student:
    school = "ABC SCHOOL"

    @classmethod
    def show_school(cls):
        print(cls.school)
```

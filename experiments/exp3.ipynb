**Experiment 3: Conditional Statements**
1.	Check whether a given number is divisible by 3 and 5 both.

```python
num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print("Number is divisible by both 3 and 5")
else:
    print("Number is NOT divisible by both 3 and 5")
```

**2.	Check whether a given number is multiple of five or not.**

```python
num = int(input("Enter a number: "))

if num % 5 == 0:
    print("Number is a multiple of 5")
else:
    print("Number is not a multiple of 5")
```

**3.	Find the greatest among two numbers. If numbers are equal than print “numbers are equal”.**

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest number is:", a)
elif b > a:
    print("Greatest number is:", b)
else:
    print("Numbers are equal")
```

**4.	Find the greatest among three numbers assuming no two values are same.**

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

if a > b and a > c:
    print("Greatest number is:", a)
elif b > a and b > c:
    print("Greatest number is:", b)
else:
    print("Greatest number is:", c)
```

**5.	Check whether the quadratic equation has real roots or imaginary roots. Display the roots.**

```python
a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

d = b*b - 4*a*c

if d > 0:
    r1 = (-b + d**0.5) / (2*a)
    r2 = (-b - d**0.5) / (2*a)
    print("Roots are real and different")
    print("Root1:", r1)
    print("Root2:", r2)
elif d == 0:
    r = -b / (2*a)
    print("Roots are real and same")
    print("Root:", r)
else:
    print("Roots are imaginary")
```

**6.	Find whether a given year is a leap year or not.**

```python
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

**8.	Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage. 
CGPA=percentage/10 
CGPA range: 
0 to 3.4 -> F 
3.5 to 5.0->C+ 
5.1 to 6->B 
6.1 to 7-> B+ 
7.1 to 8-> A 
8.1 to 9->A+ 
9.1 to 10-> O (Outstanding) 
 
Sample Gradesheet 
Name: Rohit Sharma 
Roll Number: R17234512 SAPID: 50005673 
Sem: 1 Course: B.Tech. CSE AI&ML 
 
Subject name: Marks 
PDS: 70 
Python: 80 
Chemistry: 90 
English: 60 
Physics: 50 
Percentage: 70% 
CGPA:7.0 
Grade:  A**

```python
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAP ID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

pds = int(input("Enter PDS marks: "))
python = int(input("Enter Python marks: "))
chem = int(input("Enter Chemistry marks: "))
eng = int(input("Enter English marks: "))
phy = int(input("Enter Physics marks: "))

total = pds + python + chem + eng + phy
percentage = total / 5
cgpa = percentage / 10

if cgpa <= 3.4:
    grade = "F"
elif cgpa <= 5.0:
    grade = "C+"
elif cgpa <= 6.0:
    grade = "B"
elif cgpa <= 7.0:
    grade = "B+"
elif cgpa <= 8.0:
    grade = "A"
elif cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O"

print("\n----- Grade Sheet -----")
print("Name:", name)
print("Roll Number:", roll)
print("SAP ID:", sapid)
print("Sem:", sem)
print("Course:", course)

print("\nSubject Marks:")
print("PDS:", pds)
print("Python:", python)
print("Chemistry:", chem)
print("English:", eng)
print("Physics:", phy)

print("Percentage:", percentage, "%")
print("CGPA:", round(cgpa, 1))
print("Grade:", grade)
```

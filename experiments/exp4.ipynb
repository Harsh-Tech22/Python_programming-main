Q1) Find a factorial of given number.

```python
n = int(input("Enter a number: "))
fact = 1

for i in range(1, n + 1):
    fact = fact * i

print("Factorial:", fact)
```

**Q2) Find whether the given number is Armstrong number.**

```python
num = int(input("Enter a number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit**3
    temp = temp // 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

**Q3) Print Fibonacci series up to given term.**

```python
n = int(input("Enter number of terms: "))

a = 0
b = 1

print("Fibonacci Series:")
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

**Q4) Write a program to find if given number is prime number or not.**

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a Prime Number")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
```

**Q5) Check whether given number is palindrome or not.**

```python
num = int(input("Enter a number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if rev == num:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

**Q6) Write a program to print sum of digits.**

```python
num = int(input("Enter a number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Sum of digits:", sum)
```

**Q7) Count and print all numbers divisible by 5 or 7 between 1 to 100.**

```python
count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nCount:", count)
```

**Q8) Convert all lower cases to upper case in a string.**

```python
s = input("Enter a string: ")
result = ""

for ch in s:
    if ch >= 'a' and ch <= 'z':
        result += chr(ord(ch) - 32)
    else:
        result += ch

print("Uppercase String:", result)
```

**Q9) Print all prime numbers between 1 and 100.**

```python
print("Prime numbers between 1 and 100:")

for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
```

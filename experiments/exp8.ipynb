1.	Add few names, one name in each row, in “name.txt file”. 
a.	Count no of names 
b.	Count all names starting with vowel 
c.	Find longest name 


```python
file = open("name.txt", "w")
file.write("Amit\n")
file.write("Rohit\n")
file.write("Anjali\n")
file.write("Neha\n")
file.write("Ishita\n")
file.close()
```

```python
file = open("name.txt", "r")
names = file.readlines()
print("Number of names:", len(names))
file.close()
```

```python
file = open("name.txt", "r")
count = 0

for name in file:
    if name[0].lower() in ['a', 'e', 'i', 'o', 'u']:
        count += 1

print("Names starting with vowel:", count)
file.close()
```

```python
file = open("name.txt", "r")
longest = ""

for name in file:
    name = name.strip()
    if len(name) > len(longest):
        longest = name

print("Longest name:", longest)
file.close()
```

**2.	Store integers in a file. 
a.	Find the max number 
b.	Find average of all numbers 
c.	Count number of numbers greater than 100**

```python
file = open("numbers.txt", "w")
file.write("45\n120\n300\n78\n150\n")
file.close()
```

```python
file = open("numbers.txt", "r")
nums = [int(x) for x in file]
print("Maximum number:", max(nums))
file.close()
```

```python
file = open("numbers.txt", "r")
nums = [int(x) for x in file]
avg = sum(nums) / len(nums)
print("Average:", avg)
file.close()
```

```python
file = open("numbers.txt", "r")
count = 0

for num in file:
    if int(num) > 100:
        count += 1

print("Numbers greater than 100:", count)
file.close()
```

**3.	Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) ): 
Example: 
Dehradun 5.78 308.20 
Delhi 190 1484 
…………… 
Open file city.txt and read to: 
a.	Display details of all cities 
b.	Display city names with population more than 10Lakhs 
c.	Display sum of areas of all cities**

```python
file = open("city.txt", "w")
file.write("Dehradun 5.78 308.20\n")
file.write("Delhi 190 1484\n")
file.write("Mumbai 124 603\n")
file.write("Pune 31 331\n")
file.write("Agra 15 188\n")
file.close()
```

```python
file = open("city.txt", "r")

print("City Details:")
for line in file:
    print(line.strip())

file.close()
```

```python
file = open("city.txt", "r")

print("Cities with population > 10 lakhs:")
for line in file:
    data = line.split()
    if float(data[1]) > 10:
        print(data[0])

file.close()
```

```python
file = open("city.txt", "r")
total_area = 0

for line in file:
    data = line.split()
    total_area = total_area + float(data[2])

print("Total area of all cities:", total_area)

file.close()
```

**4.	Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.  
Sample input 
1 0 
2 $ 
3 1  
Sample Output: 
Error Code: integer division or modulo by zero  
Error Code: invalid literal for int() with base 10: '$' 3**

```python
n = int(input("Enter number of test cases: "))

for i in range(n):
    try:
        a, b = input("Enter a and b: ").split()
        a = int(a)
        b = int(b)
        print(a // b)
    except ZeroDivisionError:
        print("Error Code: integer division or modulo by zero")
    except ValueError as e:
        print("Error Code:", e)
```

**5.	Create multiple suitable exceptions for a file handling program.**

```python
class FileNotFound(Exception):
    pass

class EmptyFile(Exception):
    pass

try:
    file = open("data.txt", "r")
    content = file.read()

    if content == "":
        raise EmptyFile("File is empty")

    print(content)
    file.close()

except FileNotFoundError:
    print("File not found")

except EmptyFile as e:
    print(e)
```

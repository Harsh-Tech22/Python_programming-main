1.	Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  (Note: Do not use built-in functions.)

```python
def find_max_min(lst):
    max_val = lst[0]
    min_val = lst[0]

    for num in lst:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    return max_val, min_val


numbers = [10, 6, 8, 90, 12, 56]
result = find_max_min(numbers)
print("Maximum:", result[0])
print("Minimum:", result[1])
```

**2.	Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.**

```python
def sum_of_cubes(n):
    sum = 0
    for i in range(1, n):
        sum += i ** 3
    return sum


n = int(input("Enter a positive number: "))
print("Sum of cubes:", sum_of_cubes(n))
```

**3.	Write a Python function to print 1 to n using recursion. (Note: Do not use loop)**

```python
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)


n = int(input("Enter n: "))
print_numbers(n)
```

**4.	Write a recursive function to print Fibonacci series upto n terms.**

```python
def fib(n):
    if n <= 1:
        return n
    return fib(n - 1) + fib(n - 2)


n = int(input("Enter number of terms: "))
for i in range(n):
    print(fib(i), end=" ")
```

**5.	Write a lambda function to find volume of cone.**

```python
volume_cone = lambda r, h: (1/3) * 3.14 * r * r * h

r = float(input("Enter radius: "))
h = float(input("Enter height: "))

print("Volume of cone:", volume_cone(r, h))
```

**6.	Write a lambda function which gives tuple of max and min from a list. 
Sample input: [10, 6, 8, 90, 12, 56] 
Sample output: (90,6)**

```python
data = [10, 6, 8, 90, 12, 56]

max_min = lambda lst: (
    max(lst),
    min(lst)
)

print(max_min(data))
```

**7.	Write functions to explain mentioned concepts: 
a.	Keyword argument 
b.	Default argument 
c.	Variable length argument**

1.Keyword argument

```python
def student(name, age):
    print("Name:", name)
    print("Age:", age)


student(age=20, name="Harsh")
```

**2.Default argument**

```python
def greet(name="Student"):
    print("Hello", name)


greet()
greet("Harsh")
```

**3. Variable length argument**

```python
def total_marks(*marks):
    sum = 0
    for m in marks:
        sum += m
    print("Total Marks:", sum)


total_marks(70, 80, 60)
total_marks(50, 65, 75, 85)
```

```python
class Rapper:
    def _init_(self, name, age, genre):
        self.name = name,
        self.age = age,
        self.genre = genre

        def to_dict(self):
            return{
                "name":self.name,
                "age":self.age,
                "genre":self.genre
            }
        name = input("Enter rapper name:")
        age = int(input("Enter age:"))
        genre = input("Enter genre")

        rapper1 = Rapper(name, age, genre)
        data = rapper1.to_dict()
        json_data = json.dumps(data, indent=4)

        print("JSON Output:")
        print(json_data)
```

```python
class Car:
    def __init__(self, name, color):
        self.name = name
        self.color = color
class brezza(Car):
    def __init__(self, name, color, seats,bumpers):
        self.name = name
        self.color = color
        self.seats = seats
        self.bumpers = bumpers
s2 = brezza("Brezza","White","5", "2")
print(s2.name)
print(s2.color)
print(s2.seats)
print(s2.bumpers)
```

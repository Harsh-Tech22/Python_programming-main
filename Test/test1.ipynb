```python
def top_k(items, k):
    items.sort(key=lambda x: x[1], reverse=True)

    return items[:k]
items = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2

print(top_k(items,k))
```

```python
def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        d.setdefault(cat, {})
        d[cat].setdefault(word, 0)
        d[cat][word] =+ 1
    return d
pairs = [("sports","win"), ("sports","win"), ("tech","ai")]
print(count_by_category(pairs))
```

question 3:

```python
def missing_number(nums):
    n = len(nums)
    total = n * (n + 1) / 2
    return int(total - sum(nums))
nums = [3, 0, 1]
print(missing_number(nums))
```

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    if x not in seen:
        seen.add(x)
    return list(seen)

seen = set()

print(add_unique(1, seen))
print(add_unique(2, seen))
print(add_unique(1, seen))
```

```python
def moving_average(stream, w):
    result = []
    window_sum = 0
    window = []

    for num in stream:
        window.append(num)
        window_sum += num

        if len(window) > w:
            window_sum -= window.pop(0)

        result.append(window_sum / len(window))

    return result
stream = [1, 2, 3, 4, 5]
w = 3

print(moving_average(stream, w))
```
